```python
dfs_combinations = pd.read_csv(r"C:\Users\Markos Alves\Downloads\multiplascompras.csv".replace("\\","/"))
dfs_combinations = dfs_combinations.drop(columns=['Unnamed: 0'])

dfs_combinations[['data', 'hora']] = dfs_combinations['Order Date'].str.split(' ', expand=True)
dfs_combinations[['mes', 'dia', 'ano']] = dfs_combinations['data'].str.split('/', expand=True)
dfs_combinations['data'] = dfs_combinations['dia']+'/'+dfs_combinations['mes']+'/'+'20'+dfs_combinations['ano']+' '+dfs_combinations['hora']
dfs_combinations = dfs_combinations.drop(columns=['hora', 'mes', 'dia', 'ano'])
dfs_combinations = dfs_combinations.reset_index()
dfs_combinations = dfs_combinations.drop(columns=['index'], axis=1)

dfs_combinations.to_csv(r"C:\Users\Markos Alves\Downloads\multiplascompras.csv".replace("\\","/"))

dfs_combinations
```

```python
data_frame_vendas_2019['Price Each'] = data_frame_vendas_2019['Price Each'].astype(float)
data_frame_vendas_2019['Quantity Ordered'] = data_frame_vendas_2019['Quantity Ordered'].astype(float)
```

```python
products = dfs_combinations['Product'].unique()
print(len(products))
print(products)
```

```python
for order_id in dfs_combinations['Order ID'].unique():
    combinacao = data_frame_vendas_2019[data_frame_vendas_2019['Order ID'] == order_id]
    combinacao = combinacao['Product'].unique()
    combinacoes.append(combinacao)
combinacoes
```

```python
#Data frame com o numedo de vezes que os produtos foram vendidos
data_frame_vendas_counts = data_frame_vendas_2019[data_frame_vendas_2019['Product'].isin(products)]
data_frame_vendas_counts = data_frame_vendas_counts.groupby(['Product']).count()
data_frame_vendas_counts = data_frame_vendas_counts.reset_index()
```

```python
data_frame_vendas_cm = data_frame_vendas_2019[data_frame_vendas_2019['Product'].isin(products)]
data_frame_vendas_cm = data_frame_vendas_cm.groupby(['Product']).mean()
data_frame_vendas_cm = data_frame_vendas_cm.reset_index()
```

```python
# Numero de vezes que os produtos foram comprados isolados
compras_unicas = r"C:\Users\Markos Alves\Downloads\comprasunicas.csv".replace("\\","/")

compras_unicas = pd.read_csv(compras_unicas)
compras_unicas = compras_unicas[compras_unicas['Product'].isin(products)]
compras_unicas = compras_unicas.groupby(['Product']).count()
compras_unicas = compras_unicas.reset_index()
```

```python
# Numero de vezes que os produtos foram comprados combinado
compras_multiplas = r"C:\Users\Markos Alves\Downloads\multiplascompras.csv".replace("\\","/")

compras_multiplas = pd.read_csv(compras_multiplas)
compras_multiplas = compras_multiplas[(compras_multiplas['Product'].isin(products)) & (compras_multiplas['Product'] != 'Product') ]
compras_multiplas = compras_multiplas.groupby(['Product']).count()
compras_multiplas = compras_multiplas.reset_index()
```

```python
final_data_frame = pd.DataFrame({
    "produto vendas":data_frame_vendas_counts['Product'],
    "N de vendas": data_frame_vendas_counts['Order ID'],
    "produto unicos": compras_unicas['Product'],
    "Vendas unicas": compras_unicas['Order ID'],
    "produto multiplos": compras_multiplas['Product'],
    "vendas multiplas": compras_multiplas['Order ID'],
    "Produto medio": data_frame_vendas_cm['Product'],
    "Valor medio": data_frame_vendas_cm["Price Each"]
})
```
