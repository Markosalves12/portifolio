```python
import pandas as pd
import os
```

```python
arquivos_CAP = r"C:\Users\MarkosPereira.exc\OneDrive - Suzano S A\Documentos\BIdopool\CAP".replace("\\","/")

for raiz, diretorios, arquivos in os.walk(arquivos_CAP):
    for arquivo in arquivos:
        print(arquivo)
```

```python
df_arq_cap = list()
for raiz, diretorios, arquivos in os.walk(arquivos_CAP):
    for arquivo in arquivos:
        caminho_completo = os.path.join(raiz, arquivo)
        df = pd.read_excel(caminho_completo, engine='openpyxl', parse_dates=True)
        df_arq_cap.append(df)
```

```python
def tratamento(df):
    df = pd.DataFrame(df)
    cabecalhos = df.iloc[1]
    rename = df.columns
    alter = dict()
    cont = 0
    for coluna in rename:
        alter[coluna]=cabecalhos[cont]
        cont+=1
    df = df.rename(columns=alter)
    index = df[df['CAP']=='CAP'].index
    df = df.drop(index=index)
    return df
```

```python
data_frames_alterado = list()
for data_frame in df_arq_cap:
    data_frame = tratamento(data_frame)
    data_frames_alterado.append(data_frame)
    print(data_frame.columns)
    print(len(data_frame.columns))
```
