### Trabalhando os dataset referentes aos candidatos

Os datasets referentes aos candidatos forma identificados e separados em 4 grandes conjuntos listados a cima
O objetivo a partide agora é combinar os datasets pertencentes aos mesmo conjunto em um unico dataframe para facilitar as analises.

Concatenando todos os dataframes referentes as **depesas contratadas pelos candidatos** de todos os condidatos em todas as UF's candidatos em um unico dataframe, e refazemos para todos os termos prinicipais dos data frames definidos anteriormente

```python
#vamos posicionar os data frames com ''despesas_contratadas_candidatos_2022'' em uma lista
df_desp_contratadas = list()
for raiz, diretorios, arquivos in os.walk(prestacao_de_contas_dos_candidatos):
    for arquivo in arquivos:
        if not 'BRASIL' or not 'BR' in arquivo:
            if not 'leiame' in arquivo:
                if 'despesas_contratadas_candidatos_2022' in arquivo:
                    caminho_completo = os.path.join(raiz, arquivo)
                    df = pd.read_csv(caminho_completo, encoding='latin-1', delimiter=';')
                    df_desp_contratadas.append(df)
```

```python
#função para converter colunas str em float
def conv_str_func(DataFrame = pd.DataFrame(), coluna = 'coluna'):
    DataFrame[f'{coluna}'] = DataFrame[f'{coluna}'].str.replace(',','.')
    DataFrame[f'{coluna}'] = DataFrame[f'{coluna}'].astype(float)
    return DataFrame
```

```python
df_PCDC_des_contratadas = pd.concat(df_desp_contratadas)
```

##### dataframe com todas as despesas contratadas

```python
# convertendo a coluna "VR_DESPESA_CONTRATADA", para permitir operações de soma.
conv_str_func(df_PCDC_des_contratadas, 'VR_DESPESA_CONTRATADA')
```

Observe que os valores ausentes estão preenchidos com '#NULO#', nesse caso é viavel contar quantos aparecem em cada coluna

### 1 - Qual o candidato que mais contratou despesas em cada estado e em todo o brasil até a data de atualização dos arquivos. 

Olhando a ultima linha acima, podemos ver que o canditado que mais gastou até a data de referencia, é o candidato rodrigo garcia gastando mais de 15 milhões de reais nas campanhas

```python
#para responder essa pergunta precisaremos da coluna, NM_CANDIDATO, VR_DESPESA_CONTRATADA, SG_PARTIDO, SG_UF 

# Data frame com as colunas necessarias para responder essa questão
df_question1 = df_PCDC_des_contratadas[['NM_CANDIDATO','VR_DESPESA_CONTRATADA','DS_CARGO','SG_PARTIDO','SG_UF']]
```

```python
Estados = df_question1['SG_UF'].unique()
print(Estados)
```

##### Candidato que mais gastou nas suas respectivas disputas em cada estado:

```python
for estado in Estados:
    print(estado)
    new_df = df_question1[df_question1['SG_UF'] == estado]
    candidatos = new_df['NM_CANDIDATO'].unique()
    cargos = new_df['DS_CARGO'].unique()
    for cargo in cargos:
        total_cargo = 0
        new_df2 = new_df[(new_df['DS_CARGO'] == cargo)]
        for candidato in candidatos:
            new_df3 = new_df2[(new_df2['NM_CANDIDATO'] == candidato)]
            total_candidato = new_df3['VR_DESPESA_CONTRATADA'].sum()
            if total_candidato > total_cargo:
                total_cargo = total_candidato
                nome = candidato
                sigla_partido = new_df3['SG_PARTIDO'].unique()
            if candidato == candidatos[-1]: 
                print(f"{cargo}---{nome}-{sigla_partido[0]}--R${total_cargo:.2f}")
    print()
```

##### Candidato que mais contratou despesas em cada estado independente da disputa

```python
for estado in Estados:
    new_df = df_question1[df_question1['SG_UF'] == estado]
    candidatos = new_df['NM_CANDIDATO'].unique()
    total_geral = 0
    for candidato in candidatos:
        new_df2 = new_df[(new_df['NM_CANDIDATO'] == candidato)]
        total_candidato = new_df2['VR_DESPESA_CONTRATADA'].sum()
        if total_candidato > total_geral:
            total_geral = total_candidato
            nome = candidato
            sigla_partido = new_df2['SG_PARTIDO'].unique()
            cargor = new_df2['DS_CARGO'].unique()
        if candidato == candidatos[-1]:
            print(f"[{estado}]{cargor[0]}--{nome}-{sigla_partido[0]}--R${total_geral:.2f}")
            print()
```

##### Candidatos que mais contrataram despesas em todas as diputas idependente do estado

```python
cargos = df_PCDC_des_contratadas['DS_CARGO'].unique()
for cargo in cargos:
    new_df = df_question1[df_question1['DS_CARGO'] == cargo]
    candidatos = new_df['NM_CANDIDATO'].unique()
    total = 0
    for candidato in candidatos:
        new_df2 = new_df[(new_df['NM_CANDIDATO'] == candidato)]
        total_candidato_cargo = new_df2['VR_DESPESA_CONTRATADA'].sum()
        if total_candidato_cargo > total:
            agente = candidato
            total = total_candidato_cargo
            partido = new_df2['SG_PARTIDO'].unique()
            estado = new_df2['SG_UF'].unique()
    print(f"O candidato que mais contratou despesas na disputa ao cargo de {cargo}\n"
          f"{agente}-{partido[0]}/{estado[0]}, contratou mais de R${total:.2f}")
    print()
```

### 4 - Candidatos que mais gastaram, para a disputa de deputado federal, estadual, senador e governador no Brasil e em cada unidade federal até a data de atualização dos arquivos.
