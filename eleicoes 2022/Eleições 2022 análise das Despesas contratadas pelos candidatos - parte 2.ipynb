### Trabalhando os dataset referentes aos candidatos

Os datasets referentes aos candidatos forma identificados e separados em 4 grandes conjuntos listados a cima
O objetivo a partide agora é combinar os datasets pertencentes aos mesmo conjunto em um unico dataframe para facilitar as analises.

Concatenando todos os dataframes referentes as **depesas contratadas** de todos os condidatos em todas as UF's candidatos em um unico dataframe, e refazemos para todos os termos prinicipais dos data frames definidos anteriormente

```python
df_desp_contratadas = list()
keys = list()
for raiz, diretorios, arquivos in os.walk(prestacao_de_contas_dos_candidatos):
    for arquivo in arquivos:
        if not 'BRASIL' or not 'BR' in arquivo:
            if not 'leiame' in arquivo:
                if 'despesas_contratadas_candidatos_2022' in arquivo:
                    caminho_completo = os.path.join(raiz, arquivo)
                    df = pd.read_csv(caminho_completo, encoding='latin-1', delimiter=';')
                    df_desp_contratadas.append(df)
```

```python
#função para converter colunas str em float
def conv_str_func(DataFrame = pd.DataFrame(), coluna = 'coluna'):
    DataFrame[f'{coluna}'] = DataFrame[f'{coluna}'].str.replace(',','.')
    DataFrame[f'{coluna}'] = DataFrame[f'{coluna}'].astype(float)
    return DataFrame

def conv_str_func_str(DataFrame = pd.DataFrame(), coluna = 'coluna'):
    DataFrame[f'{coluna}'] = DataFrame[f'{coluna}'].astype(str)
    return DataFrame
```

```python
df_PCDC_des_contratadas = pd.concat(df_desp_contratadas)
conv_str_func(df_PCDC_des_contratadas, coluna = 'VR_DESPESA_CONTRATADA')
conv_str_func_str(df_PCDC_des_contratadas, coluna = 'NR_CPF_CNPJ_FORNECEDOR')
conv_str_func_str(df_PCDC_des_contratadas, coluna = 'NR_CNPJ_PRESTADOR_CONTA')
```

```python
x = list()
y = list()
for coluna in df_PCDC_des_contratadas.columns:
    x.append(coluna)
    cont = 0
    for value in df_PCDC_des_contratadas[f'{coluna}']:
        if value == '#NULO#':
            cont += 1
            
    y.append((cont/df_PCDC_des_contratadas.shape[0])*100)
    
frame = {'COLUNA':x, '% de #NULO#s':y}
frame = pd.DataFrame(frame)
display(frame.sort_values(['% de #NULO#s'], ascending = False))
```

As colunas 'DS_DESPESA', 'NM_FORNECEDOR', 'DS_TIPO_DOCUMENTO', 'NR_DOCUMENTO', 'NM_FORNECEDOR_RFB' possui poucos valores ausentes, já a coluna 'DS_CNAE_FORNECEDOR' possui bastantes valores ausentes por se tratar de um tipo de documento (CNAE) pouco usado por empresas, por ja estar presente no cnpj, equanto as colunas 'SG_PARTIDO_FORNECEDOR', 'NM_PARTIDO_FORNECEDOR', 'SG_UF_FORNECEDOR', 'DS_CARGO_FORNECEDOR', 'NM_MUNICIPIO_FORNECEDOR', 'DS_ESFERA_PART_FORNECEDOR', possuem muitos valores ausentes, portanto deve existir um cuidado maior ao fazer qualquer operação com essas colunas.
