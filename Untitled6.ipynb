```python
import pandas as pd
import os
import numpy as np
```

```python
caminho_arquivos = r"C:\Users\Markos Alves\Downloads\sales 2019".replace("\\","/")

for raiz, diretorios, arquivos in os.walk(caminho_arquivos):
    for arquivo in arquivos:
        print(arquivo)
```

```python
df_arq_cap = list()
for raiz, diretorios, arquivos in os.walk(caminho_arquivos):
    for arquivo in arquivos:
        caminho_completo = os.path.join(raiz, arquivo)
        df = pd.read_csv(caminho_completo)
        df_arq_cap.append(df)
```

```python
data_frame_vendas_2019 = pd.concat(df_arq_cap)
```

```python
data_frame_vendas_2019[['data', 'hora']] = data_frame_vendas_2019['Order Date'].str.split(' ', expand=True)
data_frame_vendas_2019[['mes', 'dia', 'ano']] = data_frame_vendas_2019['data'].str.split('/', expand=True)
data_frame_vendas_2019['data'] = data_frame_vendas_2019['dia']+'/'+data_frame_vendas_2019['mes']+'/'+'20'+data_frame_vendas_2019['ano']+' '+data_frame_vendas_2019['hora']
data_frame_vendas_2019 = data_frame_vendas_2019.drop(columns=['hora', 'mes', 'dia', 'ano'])
data_frame_vendas_2019 = data_frame_vendas_2019.reset_index()
data_frame_vendas_2019 = data_frame_vendas_2019.drop(columns=['index'], axis=1)
index_to_drop = data_frame_vendas_2019[data_frame_vendas_2019['Order ID'].isnull()].index
data_frame_vendas_2019 = data_frame_vendas_2019.drop(index=index_to_drop)
```
